```python
# Exercise
# Find the Cheapest Item 
# Try to remove RM
    
from bs4 import BeautifulSoup
import re

# Assuming `soup` is your BeautifulSoup object containing the parsed HTML

soup_list = soup.findAll("li", attrs={"class": "s-item s-item__pl-on-bottom"})

cheapest = []

for item in soup_list:
    postage_span = item.find("span", class_="s-item__price")
    
    if postage_span:
        postage_cost_str = postage_span.text.strip()
        # Extract the numeric value
        #print(postage_cost_str)
        match = re.search(r'RM (\d+\.\d+)', postage_cost_str)
        #print(match)
        if match:
            postage_cost = float(match.group(1))
            # print(postage_cost)
            cheapest.append(postage_cost)

if cheapest:
    print("Cheapest Item is: RM", min(cheapest))
else:
    print("No items with postage cost found")
```

### Exercise

Store the results obtained into a Dataframe and save it as CSV file

```python
import pandas as pd
# Title column : product_name, product_price

soup_list = soup.findAll("li", attrs={"class":re.compile(r"sresult lvresult clearfix li[ shic]*")})
product_name = []
product_price = []
for items in soup_list:
    product_name.append(items.find("h3", attrs={"class":"lvtitle"}).text)
    product_price.append(float(re.sub(r'\nRM ','', items.find("span", attrs={"class":"bold"}).text)))

data = {'product_name':product_name,
       'product_price':product_price}

df_xiaomi = pd.DataFrame(data)
df_xiaomi

df_xiaomi.to_csv(r'Products.csv', index=False)
```
