**Notes:**

Python is called an "object-oriented programming language." 

* An object in python means collection of data and functions that act on those data.

* This means there is a construct in Python called a class that lets you structure your software in a particular way. Using classes, you can add consistency to your programs so that they can be used in a cleaner way also known as abstraction.

* A class is a way to take a grouping of functions and data and place them inside a container and you can access them with the . (dot) operator. 

* You can create multiple instances of a class, and each one won't interfere with each other. The result is called an object and you then assign it to a variable to work with it.

* In the following cell, we are going to create an blueprint that can let us create an object "human".



```python
class Human:
#   Here is where we define the properties the object to have  
    def __init__(self, name, age, height, hometown): # constructor
        self.name = name
        self.age = age
        self.height = height
        self.hometown = hometown

#   Then we define what actions the object to have
    def introduce(self):
        print("Hello! My name is %s" %self.name)
        print("I am %d years old" %self.age)
        print("I am %d cm tall" %self.height)
        print("My hometown is %s" %self.hometown)

#  Initialize object Human
daren = Human('Daren Tan',28, 183, 'Ipoh')

# Call values in object
print(daren.name)

# Call function in object
daren.introduce()
```

```python
class Animal:
  def __init__(self, animal, name, age, color, legs):
    self.animal = animal
    self.name = name
    self.age = age
    self.color = color
    self.legs = legs

  def sound(self):
    if(self.animal == "Dog"):
        print("woof")
    elif(self.animal == "Cat"):
        print("meow")
    else:
        print("wapapapow")

dog = Animal("Dog", "BB", 10, "Brown", 4)
dog.sound()
dog.color = "yellow"
print(dog.color)

del dog.color
print(dog.color) # will give you an error

cat = Animal("Cat", "Lucy", 5, "Black", 4)
cat.sound()

fox = Animal("Fox", "John", 3, "Red", 4)
fox.sound()

del fox
fox.sound() # will give you an error
```

```python
# to showcase object duplication and how it affects the object's values
dog = Animal("Dog", "BB", 10, "Brown", 4)
dog_2 = dog 
print("first dog's id is:", id(dog_2))
print("second dog's id is:", id(dog))

print("first dog's name is: ", dog.name)
dog_2.name = "CC"
print("first dog's name after changed is: ", dog.name)

import copy
print("-------now I make it a copy--------")
dog_2 = copy.copy(dog)
print("first dog's id is:", id(dog_2))
print("second dog's id is:", id(dog))

print("first dog's name is: ", dog.name)
dog_2.name = "AA"
print("first dog's name after changed is: ", dog.name)


```

```python
# Practical Example
# Taken fromhttps://jeffknupp.com/blog/2014/06/18/improve-your-python-python-classes-and-object-oriented-programming/
class Customer(object):
    """A customer of ABC Bank with a checking account. Customers have the
    following properties:

    Attributes:
        name: A string representing the customer's name.
        balance: A float tracking the current balance of the customer's account.
    """

    def __init__(self, name):
        """Return a Customer object whose name is *name*.""" 
        self.name = name

    def set_balance(self, balance=0.0):
        """Set the customer's starting balance."""
        self.balance = balance

    def withdraw(self, amount):
        """You're looking at "withdraw" function: Return the balance remaining after withdrawing *amount* dollars."""
        if amount > self.balance:
            raise RuntimeError('Amount greater than available balance.')
        self.balance -= amount
        return self.balance

    def deposit(self, amount):
        """Return the balance remaining after depositing *amount*
        dollars."""
        self.balance += amount
        return self.balance
```

```python
customer = Customer('Daren')

# Check available functions for an object
print(dir(customer))

# Find out what object do
print(customer.__doc__)

# Find out what object functions do
print(customer.withdraw.__doc__)

```
